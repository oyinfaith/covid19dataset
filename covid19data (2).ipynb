```python
#I decided to remove all NAN values in the column because there are one NaN values in almost all the column which is really not significant and wont affect the work.
df = df.dropna(axis=0)
df.isnull().sum()
```

```python
#Renaming the columns for easy access and readability.
df.rename(columns={'Name':'Name','WHO Region':'Region', 'Cases - cumulative total':'Cases cum total','Cases - cumulative total per 100000 population':'Cases cum total per 100k pop','Cases - newly reported in last 7 days':'Cases newly reported in lst 7days','Cases - newly reported in last 7 days per 100000 population':'Cases newly reported in lst 7days per 100k pop','Cases - newly reported in last 24 hours':'Cases newly reported in last 24 hurs','Deaths - cumulative total':'Deaths cum total','Deaths - cumulative total per 100000 population':'Deaths cum total per 100k pop','Deaths - newly reported in last 7 days':'Deaths newly reported in lst 7days','Deaths - newly reported in last 7 days per 100000 population':'Deaths newly reported in lst 7days per 100k pop','Deaths - newly reported in last 24 hours':'Deaths newly reported in lst 24 hurs'}, inplace=True)
#cum- cumulative, Pop-Population, 100k-100000, hurs-hours
```

```python
#To calc the difference btw Cases cum total and Cases cum total per 100k pop
def cal(to_calc_df):
    return to_calc_df["Cases cum total"] - to_calc_df["Cases cum total per 100k pop"]
    
df["Diff_casesOccurrence"] = df.loc[:, "Cases cum total":"Cases cum total per 100k pop"].apply(cal, axis=1)

df[['Cases cum total','Cases cum total per 100k pop', 'Diff_casesOccurrence']].head()
```

```python
#To calc the difference btw Cases reported in lst 7days and Cases reported in lst 7days per 100k pop
def cal(to_calc_df):
    return to_calc_df["Cases newly reported in lst 7days"] - to_calc_df["Cases newly reported in lst 7days per 100k pop"]
    
df["Diff_casesOccurrenceInLst7days"] = df.loc[:, "Cases newly reported in lst 7days":"Cases newly reported in lst 7days per 100k pop"].apply(cal, axis=1)

df[['Cases newly reported in lst 7days','Cases newly reported in lst 7days per 100k pop', 'Diff_casesOccurrenceInLst7days']].head()
```

```python
def calculate(to_calc_df):
    return to_calc_df["Deaths cum total"] - to_calc_df["Deaths cum total per 100k pop"]
    
df["Diff_deathcases"] = df.loc[:, "Deaths cum total":"Deaths cum total per 100k pop"].apply(calculate, axis=1)
```

```python
coviddth = df[['Deaths cum total', 'Deaths cum total per 100k pop','Diff_deathcases']].head()
coviddth
```

```python
def calc(to_calc_df):
    return to_calc_df["Deaths newly reported in lst 7days"] - to_calc_df["Deaths newly reported in lst 7days per 100k pop"]
    
df["Diff_deathcasesInLst7days"] = df.loc[:, "Deaths newly reported in lst 7days":"Deaths newly reported in lst 7days per 100k pop"].apply(calc, axis=1)

df[['Deaths newly reported in lst 7days', 'Deaths newly reported in lst 7days per 100k pop', 'Diff_deathcasesInLst7days']].head()
```
